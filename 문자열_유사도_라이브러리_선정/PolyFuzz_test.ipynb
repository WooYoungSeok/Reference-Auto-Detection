```python
import pandas as pd

df = pd.read_csv('2008_2022_한국현대문학_인문_복합학_추가데이터_참고문헌_박완서.xlsx - Sheet1.csv')
df.head()
```

```python
nan_count = df['title'].isnull().sum()
print(f"Number of NaN values in the 'title' column: {nan_count}")
```

```python
from polyfuzz import PolyFuzz
from polyfuzz.models import TFIDF

# 모델 초기화
model = PolyFuzz(TFIDF())

df = df.dropna(subset=['title'])

from_list = df['title'].tolist()
to_list = df['title'].tolist()

model.match(from_list, to_list)

# 결과 확인
matches = model.get_matches()
matches.head()
```

```python
matches[(matches['Similarity'] != 0.0) & (matches['Similarity'] != 1.0)]
```

```python
# 그룹화 기준 설정
grouped_titles = matches.groupby('From')['To'].apply(list)
print(grouped_titles)

# 중복된 제목을 그룹화된 제목으로 대체
for index, row in df.iterrows():
    if row['title'] in grouped_titles:
        df.at[index, 'grouped_title'] = grouped_titles[row['title']][0]
```
